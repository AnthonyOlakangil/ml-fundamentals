```python
import seaborn as sns
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(tips['tip'])
```

```python
sns.histplot(tips['tip'])
```

```python
sns.lineplot(x = 'size', y = 'total_bill', data = tips)
```

```python
sns.scatterplot(data = tips, x = 'tip', y = 'total_bill')
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg')
```

```python
sns.pairplot(tips, hue = 'smoker')
```

```python
sns.barplot(x = 'sex', y = 'tip', data = tips, estimator = np.std)
```

```python
sns.countplot(x = 'sex',  data = tips)
```

```python
sns.boxplot(x = 'sex', y ='total_bill', data = tips)
```

```python
sns.violinplot(x = 'smoker', y = 'tip', data = tips)
```

```python
sns.stripplot(x = 'day', y = 'tip', data = tips)
```

```python
sns.swarmplot(x = 'day', y = 'tip', data = tips)
```

```python
sns.catplot(x = 'time', y = 'tip', data = tips, )
```
