```
import random
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import numpy as np # Import numpy for handling potential empty arrays

# --- 1. Define Intents and Training Data ---
# This is where you list example phrases for each intent your chatbot should recognize.
# More examples per intent generally lead to better accuracy.
training_data = [
    ("hello", "greet"),
    ("hi there", "greet"),
    ("hey", "greet"),
    ("greetings", "greet"),
    ("good morning", "greet"),
    ("good afternoon", "greet"),
    ("good evening", "greet"),
    ("how are you doing", "ask_how_are_you"),
    ("how's it going", "ask_how_are_are_you"),
    ("what's up", "ask_how_are_you"),
    ("how do you feel", "ask_how_are_you"),
    ("are you okay", "ask_how_are_you"),
    ("tell me about yourself", "ask_name"), # Reusing ask_name for identity questions
    ("who are you", "ask_name"),
    ("what is your name", "ask_name"),
    ("your name", "ask_name"),
    ("introduce yourself", "ask_name"),
    ("bye", "goodbye"),
    ("goodbye", "goodbye"),
    ("see you later", "goodbye"),
    ("exit", "goodbye"),
    ("quit", "goodbye"),
    ("end chat", "goodbye"),
    ("thanks", "thanks"),
    ("thank you", "thanks"),
    ("appreciate it", "thanks"),
    ("thank you very much", "thanks"),
    ("thanks a lot", "thanks"),
    ("what can you do", "capabilities"),
    ("help me", "capabilities"),
    ("what are your functions", "capabilities"),
    ("what do you do", "capabilities"),
    ("can you help", "capabilities"),
    ("tell me a joke", "tell_joke"), # New intent
    ("make me laugh", "tell_joke"),
    ("got any jokes", "tell_joke"),
    ("tell me something funny", "tell_joke"),
    ("what is the weather like", "get_weather"), # New intent
    ("weather today", "get_weather"),
    ("is it raining", "get_weather"),
    ("temperature outside", "get_weather"),
    ("what's the weather", "get_weather"),
]

# Separate texts and labels (intents)
training_texts = [text for text, intent in training_data]
training_intents = [intent for text, intent in training_data]

# --- 2. Prepare Data for ML Model ---
# Use TF-IDF Vectorizer to convert text data into numerical features.
# TF-IDF stands for Term Frequency-Inverse Document Frequency.
# It reflects how important a word is to a document in a collection.
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(training_texts)

# --- 3. Train the ML Model ---
# We'll use a simple Logistic Regression classifier.
# It's a good baseline for text classification tasks.
model = LogisticRegression()
model.fit(X_train, training_intents)

# --- 4. Define Responses for Each Intent ---
# Map each intent to a list of possible responses.
responses = {
    "greet": ["Hello!", "Hi there!", "Hey!", "Greetings!"],
    "ask_how_are_you": ["I'm just a program, but I'm doing well!", "Functioning as expected!", "All good here!", "Pretty good, thanks for asking!"],
    "ask_name": ["I don't have a name.", "I am a chatbot.", "You can call me Chatbot.", "I'm a simple AI."],
    "goodbye": ["Goodbye!", "See you!", "Bye for now!", "Talk to you later!", "Farewell!"],
    "thanks": ["You're welcome!", "No problem.", "Glad I could help.", "Anytime!"],
    "capabilities": ["I can chat with you using simple responses based on what I understand.", "I can respond to basic greetings and questions.", "I can recognize a few simple intents and respond accordingly."],
    "tell_joke": ["Why don't scientists trust atoms? Because they make up everything!", "Did you hear about the mathematician who’s afraid of negative numbers? He’ll stop at nothing to avoid them!", "Why did the scarecrow win an award? Because he was outstanding in his field!"], # Responses for new intent
    "get_weather": ["I cannot check the weather right now, but I can chat!", "My apologies, I don't have access to real-time weather data.", "That's beyond my current capabilities."], # Responses for new intent
    "fallback": ["I'm not sure I understood that.", "Could you try rephrasing?", "Let's talk about something else.", "Okay.", "Hmm, I don't have a response for that."], # Fallback response if intent is unclear
}

# --- 5. Create the Chat Function ---
def get_intent(user_input):
    """
    Predicts the intent of the user's input using the trained model.
    """
    # Transform the user input using the *same* vectorizer trained on the data
    X_test = vectorizer.transform([user_input])

    # Predict the intent
    # model.predict returns an array, so we take the first element
    predicted_intent = model.predict(X_test)[0]

    # Optional: Get prediction probabilities to set a confidence threshold
    # probabilities = model.predict_proba(X_test)
    # max_proba = np.max(probabilities)
    # if max_proba < 0.5: # Example threshold
    #     return "fallback"

    return predicted_intent

def get_response(intent):
    """
    Gets a random response based on the predicted intent.
    Uses a fallback if the intent is not found in the responses dictionary.
    """
    return random.choice(responses.get(intent, responses["fallback"]))

def chat():
    """
    Runs the main chat loop using the ML model for intent recognition.
    """
    print("Hello! I'm an ML-powered chatbot. Type 'bye' to exit.")

    while True:
        user_input = input("You: ")

        if user_input.lower() in ['bye', 'goodbye', 'see you later', 'exit', 'quit', 'end chat']:
            # Predict intent for 'bye' or similar phrases to get the correct goodbye response
            # This makes the exit more robust
            intent = get_intent(user_input)
            print("Chatbot:", get_response(intent))
            break # Exit the loop

        # Get the predicted intent for the user's input
        intent = get_intent(user_input)

        # Get the response based on the predicted intent
        response = get_response(intent)

        print("Chatbot:", response)

# --- 6. Start the Chat ---
if __name__ == "__main__":
    # Ensure you have scikit-learn installed: pip install scikit-learn
    # For more advanced text processing, consider NLTK or spaCy:
    # pip install nltk spacy
    # python -m spacy download en_core_web_sm # Download English model for spaCy
    chat()
```
